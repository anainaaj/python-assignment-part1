## Task-2 Marks Analysis Using Loops & Conditionals 

```python
#Provided Data
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

```python
print(f"{student_name}'s Subject- wise Performance:")
for subject, mark in zip(subjects, marks):
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"
    
    print(f"{subject}: {mark} - Grade: {grade}")
```

```python
#calculating total marks
total_marks = sum(marks)

#calculating average marks rounded to 2 decimal places
average_marks = round(total_marks / len(marks), 2)

#finding the highest scoring subject
highest_mark = max(marks)
highest_subject = subjects[marks.index(highest_mark)]

#Finding the lowest scoring subject
lowest_mark = min(marks)
lowest_subject = subjects[marks.index(lowest_mark)]


print(f"\n{student_name}'s Overall Performance:")
print(f"Total Marks: {total_marks}")
print(f"Average Marks: {average_marks}")
print(f"Highest Subject: {highest_subject} with {highest_mark} marks")
print(f"Lowest Subject: {lowest_subject} with {lowest_mark} marks")
```

```python
#Using a while loop to simulate a marks-entry system that allows adding new subjects
new_subjects = []
new_marks = []          
while True:
    new_subject = input("Enter a new subject (or type 'done' to finish): ")
    if new_subject.lower() == 'done':
        break
    new_mark = int(input(f"Enter marks for {new_subject}: ")) #new marks input between 0 and 100
    if new_mark < 0 or new_mark > 100:
        print("Invalid marks. Please enter a value between 0 and 100.")
        continue

    new_subjects.append(new_subject)
    new_marks.append(new_mark)

    #print number of new subjects added and the updated average across all subjects (original + new)
    total_subjects = len(subjects) + len(new_subjects)
    total_marks = sum(marks) + sum(new_marks)
    average = total_marks / total_subjects if total_subjects > 0 else 0
    print(f"Number of new subjects added: {len(new_subjects)}")
    print(f"Updated average across all subjects: {average:.2f}")
```

```python
##Task 3 -Class Performance Summary 
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

```python
print("\nClass Performance Report:")
print(f"{'Student Name':<20} | {'Average':<15}|{'Status':<10}")
print("-" * 45) 
for name, marks in class_data:
    average = round(sum(marks) / len(marks), 2) #to calculate average marks rounded to 2 decimal places
    result = "Pass" if average >= 60 else "Fail" #to assign result status based on average marks
    print(f"{name:<20} | {average:<15} | {result:<10}")


pass_count   = 0
fail_count   = 0
all_averages = []     
topper_name  = ""
topper_avg   = 0

for name, marks in class_data:
    average = round(sum(marks) / len(marks), 2) #to calculate average marks rounded to 2 decimal places
    result = "Pass" if average >= 60 else "Fail" #to assign result status based on average marks
    print(f"{name:<20} | {average:<15} | {result:<10}")

 # Track pass/fail counts
    if result == "Pass":
        pass_count += 1
    else:
        fail_count += 1

    all_averages.append(average)

    # Track the class topper
    if average > topper_avg:
        topper_avg  = average
        topper_name = name

print(f"\nClass Summary:")
print(f" Students passed : {pass_count}")
print(f"  Students failed : {fail_count}")
print(f"  Class topper    : {topper_name} ({topper_avg})")
print(f"  Class average   : {average}")
```

Task 4 — String Manipulation Utility

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "
```

```python

# STEP 1: Strip whitespace from both ends
clean_essay = essay.strip()
print(f"\nStep 1 — Stripped:\n  {clean_essay}")

# STEP 2: Title Case
print(f"\nStep 2 — Title Case:\n  {clean_essay.title()}")

# STEP 3: Count occurrences of "python"
python_count = clean_essay.count("python")
print(f"\nStep 3 — 'python' appears {python_count} time(s)")

# STEP 4: Replace "python" with "Python 🐍"
replaced = clean_essay.replace("python", "Python 🐍")
print(f"\nStep 4 — After replace:\n  {replaced}")

# STEP 5: Split into sentences
sentences = clean_essay.split(". ")
print(f"\nStep 5 — Sentences list:\n  {sentences}")

# STEP 6: Print sentences numbered ending with a period
print(f"\nStep 6 — Numbered sentences:")
for i, sentence in enumerate(sentences, start=1):
    #To Check if the sentence already ends with "." to avoid double punctuation
    if not sentence.endswith("."):
        sentence = sentence + "."
    print(f"  {i}. {sentence}")   
```
